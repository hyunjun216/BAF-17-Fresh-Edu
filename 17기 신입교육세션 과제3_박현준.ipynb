```python
X = df[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age']] 

Y = df['Outcome']
```

```python
#  특정 컬럼의 '0' 값을 NaN으로 처리 
# Glucose, BloodPressure, SkinThickness, Insulin, BMI는 0이 될 수 없는 생체 측정값이므로
# 따라서 0은 결측치를 의미할 가능성이 높으므로 NaN으로 바꾼다.
zero = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

for col in zero:
    X[col] = X[col].replace(0, np.nan)

print(X.isnull().sum()[X.isnull().sum() > 0])
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42,stratify=Y) 
# 테스트 비율을 0.2로 설정
# stratify=Y 계층화 : 클래스 비율이 원본 데이터의 클래스 비율과 도일하게 유지 -> 모델의 성능 평가 신뢰도 up 

```

```python
print(X.shape) # 전체
print(X_train.shape) # 
print(X_test.shape) # 
print(Y_train.shape) # 
print(Y_test.shape) # 
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import GradientBoostingClassifier

# 전처리 및 모델 학습 과정을 하나의 파이프라인으로 연결
model_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),  # 1. 빠진 값은 중간값으로 채우고
    ('scaler', StandardScaler()),                    # 2. 값들의 크기를 통일시키고
    ('classifier', GradientBoostingClassifier(random_state=42)) # 3. 이 모델로 예측한다
])

 
```

```python
model_pipeline.fit(X_train, Y_train)
 
```

```python
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, classification_report
# 필요한 평가 지표 함수들을 불러옵니다.
 
Y_pred = model_pipeline.predict(X_test)
# 테스트 데이터(X_test)에 대한 모델의 예측 클래스(0 또는 1) 

Y_pred_proba = model_pipeline.predict_proba(X_test)[:, 1]
# 테스트 데이터에 대한 클래스 1(양성)의 예측 확률 

accuracy = accuracy_score(Y_test, Y_pred)
# 실제 값(Y_test)과 예측 값(Y_pred)을 비교하여 정확도 

f1 = f1_score(Y_test, Y_pred)
# 실제 값(Y_test)과 예측 값(Y_pred)을 비교하여 F1 스코어 

roc_auc = roc_auc_score(Y_test, Y_pred_proba)
# 실제 값(Y_test)과 예측 확률(Y_pred_proba)을 비교하여 ROC AUC 스코어 

print("\n--- Gradient Boosting 모델 평가 결과 (테스트 세트) ---")
print(f"Accuracy: {accuracy:.4f}") 
print(f"F1 Score: {f1:.4f}")
print(f"ROC AUC: {roc_auc:.4f}")
print("Classification Report:")
print(classification_report(Y_test, Y_pred))
 
```
